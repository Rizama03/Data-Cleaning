```python
# -- for the "STARS" and "DIRECTOR" columns

pattern = r'[^|]*\|' # regex pattern to match all characters up to and including the pipe symbol (|) 
# loops through each cells in "STARS" and "DIRECTOR" columns simultaneously using the same index
for index, row in df.loc[:, ["STARS", "DIRECTOR"]].iterrows():
    match = re.search(pattern, row["STARS"]) # searches for the defined pattern earlier and creates a match if found
    if match: # checks to see if the cell content of "STARS" has any characters that matches the regex pattern
        matched_content = match.group().strip('|').strip() # extracts matched content
        df.at[index, "DIRECTOR"] = matched_content  # replaces the cell content of "DIRECTOR" column with matched characters"

        df.at[index, "STARS"] = re.sub(pattern, "", row["STARS"]).strip(', ') # deletes the matched characters from the "STARS" column
 
df # confirms to see if code has been implemented
```

```python
pattern = r'[^:]*\:' # regex pattern to match all characters up to and including the colon symbol (:) 
# loops through each cells in "STARS" columns using the index
for index, row in df.loc[:, ["STARS"]].iterrows():
    df.at[index, "STARS"] = re.sub(pattern, "", row["STARS"]).strip(', ') # deletes the matched characters from the "STARS" column
 
df # confirms to see if code has been implemented
```

```python
pattern = r'^\n+$' # pattern to match all instances of new_line characters 
df["STARS"] = df["STARS"].replace(pattern, "Not specified", regex = True) # removes all instances of new_line characters 
df["STARS"] # confirms to see if code has been implemented on a row(s) that was initially affected
```
