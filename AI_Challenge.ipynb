```python
# import and read excel file
file_path = 'final_work.xlsx'
d_f = pd.read_excel(file_path)

# display dataframe - d_f
d_f
```

```python
# create duplicate for the dataframe to prevents unwanted mistakes on main excel dataframe
df = d_f.drop_duplicates()

# display duplicated dataframe
df
```

```python
# --  for the "mobile_name" column

# stripes off unwanted string from the content of every cell (i.e at the end) and saves it to the dataframe 
df['mobile_name'] = df['mobile_name'].str.rstrip('+ 8GB')

# replaces unwanted string with whitespace using regex notation and saves it to the dataframe
df['mobile_name'] = df['mobile_name'].replace(r'RAM \+\d+', '', regex=True)

# print(df.iloc[[3]])

```

```python
# -- for the "specs_score" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe
df['specs_score'] = df['specs_score'].apply(lambda x: str(x))

# replaces all "Nan" values in every cell with 0
df['specs_score'] = df['specs_score'].fillna('0', inplace=True)

# displays dataframe
df

# writes dataframe to excel document
df.to_excel('modified3.xlsx')
```

```python
# -- for the "processor" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe
df['processor'] = df['processor'].apply(lambda x: str(x))

# import regular expression library
import re

# loops through the column and places a hyphen in between an alphabet (upper or lower case) and a digit and saves it to the dataframe
df['processor'] = df['processor'].apply(lambda x:  re.sub(r'([a-zA-Z])(\d)', r'\1-\2', x))

# loops through the column and places a hyphen in between a digit and an alphabet (upper or lower case) and saves it to the dataframe
df['processor'] = df['processor'].apply(lambda x:  re.sub(r'(\d)([a-zA-Z])', r'\1-\2', x))

# writes dataframe to excel document
# df.to_excel('modified3.xlsx')
```

```python
# -- for the "storage" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe
df['storage'] = df['storage'].apply(lambda x: str(x))

# loops through the column and places a " + " between a digit followed by an alphanet AND an alphabet followed by a digit and saves it to the dataframe
df['storage'] = df['storage'].apply(lambda x:  re.sub(r'(\d)([a-zA-Z])(\d)([a-zA-Z])', r'\1\2 + \3\4', x))

# loops through the column and places a " + " between an alphabet followed by a digit and saves it to the dataframe
df['storage'] = df['storage'].apply(lambda x:  re.sub(r'([a-zA-Z])(\d)', r'\1 + \2', x))

# loops through the column and places a whitespace between a digit followed by an alphanet and saves it to the dataframe
df['storage'] = df['storage'].apply(lambda x:  re.sub(r'(\d)([a-zA-Z])', r'\1 \2', x))

# replaces all whitespaces with no space
df['storage'] = df['storage'].replace(r' ', '', regex=True)

# loops through the column and places a whitespace between the string "GB" AND an alphabet and saves it to the dataframe
df['storage'] = df['storage'].apply(lambda x:  re.sub(r'(GB)([a-zA-Z])', r'\1 \2', x))

# writes dataframe to excel document
# df.to_excel('modified3.xlsx')
```

```python
# -- for the "battery" column

# loops through the column and places a whitespace between the string "mAh" AND an alphabet and saves it to the dataframe
df['battery'] = df['battery'].apply(lambda x:  re.sub(r'(mAh)([a-zA-Z])', r'\1 \2', x))

# loops through the column and places a whitespace between the string "Battery" AND an alphabet and saves it to the dataframe
df['battery'] = df['battery'].apply(lambda x:  re.sub(r'(Battery)([a-zA-Z])', r'\1 \2', x))

# loops through the column and places a whitespace between the string "with" AND a digit and saves it to the dataframe
df['battery'] = df['battery'].apply(lambda x:  re.sub(r'(with)(\d)', r'\1 \2', x))

# displays dataframe
df
```

```python
# -- for the "display" column

# loops through the column and places a whitespace between a digit AND an alphabet and saves it to the dataframe
df['display'] = df['display'].apply(lambda x:  re.sub(r'(\d)([a-zA-Z])', r'\1 \2', x))

# loops through the column and places a whitespace between an alphabet AND a digit and saves it to the dataframe
df['display'] = df['display'].apply(lambda x:  re.sub(r'([a-zA-Z])(\d)', r'\1 \2', x))

# loops through the column and places the string " || "  between the string "inches" AND a whitespace and saves it to the dataframe
df['display'] = df['display'].apply(lambda x:  re.sub(r'(inches)(\s)', r'\1 || \2', x))

# loops through the column and places the string " || "  between the string "px" AND a whitespace and saves it to the dataframe
df['display'] = df['display'].apply(lambda x:  re.sub(r'(px)(\s)', r'\1 || \2', x))

# displays dataframe
df
```

```python
# -- for the "camera" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['camera'] = df['camera'].apply(lambda x: str(x))

# loops through the column and places a whitespace between the string "dp" AND a digit and saves it to the dataframe
df['camera'] = df['camera'].apply(lambda x:  re.sub(r'(MP)(\d)', r'\1 \2', x))

# loops through the column and places a whitespace between the string "MP" AND an alphanet and saves it to the dataframe
df['camera'] = df['camera'].apply(lambda x:  re.sub(r'(MP)([a-zA-Z])', r'\1 \2', x))

# loops through the column and places a whtespace between an alphabet followed by a digit and saves it to the dataframe
df['camera'] = df['camera'].apply(lambda x:  re.sub(r'([a-zA-Z])(\d)', r'\1 \2', x))

# displays dataframe
df
```

```python
# -- for the "extra_storage" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['extra_storage'] = df['extra_storage'].apply(lambda x: str(x))

# loops through the column and places a whitespace between the strings "Memory" AND "Card" and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].apply(lambda x:  re.sub(r'(Memory)(Card)', r'\1 \2', x))

# loops through the column and places a whitespace between the strings "Card" AND "Not" and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].apply(lambda x:  re.sub(r'(Card)(Not)', r'\1 \2', x))

# loops through the column and places a whitespace between the strings "Not" AND "Supported" and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].apply(lambda x:  re.sub(r'(Not)(Supported)', r'\1 \2', x))

# loops through the column and places a whitespace between an alphabet AND a whitespace and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].apply(lambda x:  re.sub(r'([a-zA-Z])(\s)', r' \2', x))

# replaces the string "Memory CardHybrid" with "Memory Card Not Supported" and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].replace(r'Memory CardHybrid', 'Memory Card Not Supported', regex=True)

# replaces the string "Memory CardSupported" with "Memory Card Not Supported" and saves it to the dataframe
df['extra_storage'] = df['extra_storage'].replace(r'Memory CardSupported', 'Memory Card Not Supported', regex=True)

# writes dataframe to excel document
df.to_excel('new2.xlsx')

# displays dataframe
df
```

```python
# -- for the "os" and "missing_features" columns

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['os'] = df['os'].apply(lambda x: str(x))

# loops through each cells in "os" and "missing_features" columns simultaneously using the same index
for index, row in df.loc[:, ["os", "missing_features"]].iterrows():
    
    if row["os"] == "No FM Radio": # checks to see if the cell content of "os" column reads "No FM Radio"
        df.at[index, "missing_features"] = df.at[index, "os"] # replaces the cell content of "missing_features" with "No FM Radio"

# displays dataframe
df

      
```

```python
# -- for the "os" and "extra_storage" columns

# loops through each cells in "os" and "extra_storage" columns simultaneously using the same index
for index, row in df.loc[:, ["os", "extra_storage"]].iterrows():
    
    if row["os"] == "No FM Radio": # checks to see if the cell content of "os" column reads "No FM Radio"
        df.at[index, "os"] = df.at[index, "extra_storage"] # replaces the cell content of "os" column with cell content of "extra_storage" using the same index 
        df.at[index, "extra_storage"] = "Memory Card Not Supported" # replaces the cell content of "extra_storage" with "Memory Card Not Supported" 

# writes dataframe to excel document
df.to_excel("new4.xlsx")
```

```python
# -- for the "camera" and "extra_storage" columns 

# loops through each cells in "camera" and "extra_storage" columns simultaneously using the same index
for index, row in df.loc[:, ["camera", "extra_storage"]].iterrows():
                             
    if row["camera"] == "Foldable": # checks to see if the cell content of "camera" column reads "Foldable"
        df.at[index, "camera"] = df.at[index, "extra_storage"] + " (Foldable)" # replaces the cell content of "camera" column with cell content of "extra_storage" together with the string " (Foldable)" using the same index 
        df.at[index, "extra_storage"] = "Memory Card Not Supported" # replaces the cell content of "extra_storage" with "Memory Card Not Supported" 

# writes dataframe to excel document
df.to_excel("new6.xlsx")
```

```python
# -- for the "camera" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['camera'] = df['camera'].apply(lambda x: str(x))

# loops through all rows in "camera" column index
for index, row in df.loc[:, ["camera"]].iterrows():
    if row["camera"] == "MemoryCardNotSupported":  # checks to see if the cell content of "camera" column reads "MemoryCardNotSupported"
        df.at[index, "camera"] = "No Camera Specification" # if true, replaces the cell content of "camera" colums with "Memory Card Not Supported"

# writes dataframe to excel document
df.to_excel("new7.xlsx")
```

```python
# -- for the "processor" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['processor'] = df['processor'].apply(lambda x: str(x))

df['processor'] =  df['processor'].str.rstrip('ore') # strips off string 'ore' at the end of all cell content in the column

# loops through each cells in "processor", "os" and "media_connectivity" columns simultaneously using the same index
for index, row in df.loc[:, ["processor", "os", "media_connectivity"]].iterrows():
    if row["processor"] == "NoWifi": # checks to see if the cell content of "processor" column reads "NoWifi"
        df.at[index, "camera"] = df.at[index, "extra_storage"] # replaces the cell content of "camera" column with cell content of "extra_storage" using the same index 
        df.at[index, "extra_storage"] = df.at[index, "os"] # replaces the cell content of "extra_storage" column with cell content of "os" using the same index 
        df.at[index, "processor"] = "Processor not Allocated" # replaces the cell content of "processor" column with "Processor not Allocated"

# writes dataframe to excel document
df.to_excel("new7.xlsx")
```

```python
# -- for the "media_connectivity" column

# loops through the column and converts all cells in column to strings assuming they're of a different type and saves it to the dataframe 
df['media_connectivity'] = df['media_connectivity'].apply(lambda x: str(x))

# loops through all rows in "media_connectivity" column index
for index, row in df.loc[:, ["media_connectivity"]].iterrows():
    if row["media_connectivity"]  == "nan": # checks to see if the cell content of "media_connectivity" column reads "nan"
        df.at[index, "camera"] = df.at[index, "extra_storage"]  # replaces the cell content of "camera" column with cell content of "extra_storage" using the same index 
        df.at[index, "extra_storage"] = df.at[index, "os"] # replaces the cell content of "extra_storage" column with cell content of "os" using the same index 
        df.at[index, "media_connectivity"] = "Nil" # replaces the cell content of "media_connectivity" column with "Processor not Allocated"
        
# writes dataframe to excel document
df.to_excel("new7.xlsx")
```
